```python
import seaborn as sns
import pandas as pd
import plotly.express as px

from logits_reader import LogitsReader

def box_plot(base_path:str, dataset):
    all_prompt_outputs = []
    for prompt_num in range(0,6): 
        path = f"{base_path}/{dataset}/prompt_{prompt_num}"
        logits_reader = LogitsReader(path=path, dataset=dataset)
        prompt_outputs = logits_reader.label_words_box_plot_data(prompt_num=prompt_num)
        all_prompt_outputs += prompt_outputs

    df = pd.DataFrame(all_prompt_outputs)
    
    fig = px.box(df, x="prompt", y="acc", color="probs", points="all")
    fig.update_layout(
        autosize=False,
        width=1000, #1600 800 1000
        height=400, #500  500 400
        font=dict(size=16),
        xaxis_title="prompt template",
        yaxis_title="accuracy",
        legend_title="",
        legend=dict(
            orientation="h",
            yanchor="bottom",
            y=1.02,
            xanchor="right",
            x=1
        )
    )

    fig.show()
    return df

    #sns.boxplot(data=df, x="acc", y="prompt", hue="probs", whis=10)
```

```python
dataset = 'qqp'
base_path = f"/home/al826/rds/hpc-work/2023/prompts/calib/search/submit/outputs/flan-t5-large"
df = box_plot(base_path, dataset)
```
